# Build product embedding library

Walks a gallery directory with the structure:
```
gallery/
  <product_name>/
    img1.jpg
    img2.png
    ...
```
For each class it runs the item detector, crops the largest OBB with zero-padding,
embeds it with ConvNeXt, averages all embeddings per class, and saves the gallery
to `models/tuned/items_classification.npy`.

````python
import sys
sys.path.insert(0, "./src/python")

from pipeline import Pipeline

pipeline = Pipeline()
````

````python
GALLERY_DIR = "./gallery"  # change to your gallery root

pipeline.product_bank.build_cache(
    gallery_dir=GALLERY_DIR,
    item_detector=pipeline._detect_items,
)
````

## Verify: spot-check a single image

````python
import cv2
from matplotlib import pyplot as plt

TEST_IMAGE = "./gallery/some_product/test.jpg"  # change to any product image

img  = cv2.imread(TEST_IMAGE)
name, score = pipeline.product_bank.lookup(img)

plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title(f"{name}  (score {score:.3f})")
plt.axis("off")
plt.show()
````
